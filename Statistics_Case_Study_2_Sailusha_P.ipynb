```python
import scipy.stats as stat
%matplotlib inline
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_z_score(x, cond, mu, sigma, title, Z_stat):
    y = stat.norm.pdf(x, mu, sigma)
    z = x[cond]
    plt.plot(x, y)
    plt.fill_between(z, 0, stat.norm.pdf(z, mu, sigma))
    plt.axvline(x=Z_Stat,color = 'k',linestyle='--',label='Z_Stat')
    plt.title(title)
    plt.show()
```

```python
#Problem Statement 2: 
 
#The college bookstore tells prospective students that the average cost of its textbooks is Rs. 52 with a standard deviation
#of Rs. 4.50. A group of smart statistics students thinks that the average cost is higher. To test the bookstore’s claim 
#against their alternative, the students will select a random sample of size 100. Assume that the mean from their 
#random sample is Rs. 52.80. Perform a hypothesis test at the 5% level of significance and state your decision.


#Solution:
# H0 -> Null Hypothesis: 𝜇 = 52
# H1 -> Alternative Hypothesis: 𝜇 > 52 #As it says students think that average cost is higher, hence a one-tail Z-test

Mean_P = 52
Stdev_p = 4.50
Mean_S = 52.80
alpha = 0.05
probability = 1 - alpha # 95% confidence interval
n = 100
Z_Stat = round((Mean_S - Mean_P)/(Stdev_p/n**0.5),2)
Z_critical = round(stat.norm.ppf(probability),2) #One-Tail Z-Test
print("Z Statistics:",Z_Stat)
print("Z critical:",Z_critical)
if Z_Stat < Z_critical:
    print("Accept Null Hypothesis i.e., average cost of textbooks is Rs.52 ")
else:
    print("Z value falls under rejection region")
    print("Reject Null Hypothesis i.e., average cost of textbooks is higher than the claim")
x = np.arange(-4,4,0.001)
l =   "z < " +  str(Z_critical)
draw_z_score(x, x < Z_critical, 0, 1, l,Z_Stat)

```

```python
#Problem Statement 3: 

#A certain chemical pollutant in the Genesee River has been constant for several years with mean μ = 34 ppm 
#(parts per million) and standard deviation σ = 8 ppm.#A group of factory representatives whose companies discharge
#liquids into the river is now claiming that they have lowered the average with improved filtration devices. 
#A group of environmentalists will test to see if this is true at the 1% level of significance. 
#Assume \ that their sample of size 50 gives a mean of 32.5 ppm. 
#Perform a hypothesis test at the 1% level of significance and state your decision. 

#Solution:
# H0 -> Null Hypothesis: 𝜇 = 34 
# H1 -> Alternative Hypothesis: 𝜇 < 34 #Performing one-tail test

Mean_P = 34
Stdev_p = 8
Mean_S = 32.50
alpha = 0.01
probability = 1 - alpha
n = 50
Z_Stat = (Mean_S - Mean_P)/(Stdev_p/n**0.5)
Z_critical = round(stat.norm.ppf(probability),2) #left sided One-Tail Test
print("Z Statistics:",Z_Stat)
print("Z critical:",Z_critical)
if Z_Stat < Z_critical:
    print("Z value falls in acceptance region")
    print("Fail to reject Null Hypothesis i.e., the chemical pollutant is 34ppm")
else:
    print("Z value falls in rejection region")
    print("Reject Null Hypothesis")
x = np.arange(-4,4,0.001)
l =  str(-Z_critical) + " < " + "z " 
draw_z_score(x, (-Z_critical < x) , 0, 1, l,Z_Stat)
```

```python
#Problem Statement 5:  
 
#In a report prepared by the Economic Research Department of a major bank the Department manager maintains that the 
#average annual family income on Metropolis is $48,432. What do you conclude about the validity of the report if a random
#sample of 400 families shows and average income of $48,574 with a standard deviation of 2000? 

#Solution:
# H0 -> Null Hypothesis: 𝜇 = 48432
# H1 -> Alternative Hypothesis: 𝜇 != 48432

Mean_P = 48432
std = 2000
Mean_S = 48574
n = 400
alpha = 0.05 # 5% significance level
prob = 1 - (alpha/2)
Z_Stat = (Mean_S - Mean_P)/(std/n**0.5)
Z_critical = round(stat.norm.ppf(prob),2)
print("Z Statistics:",Z_Stat)
print("Z critical:",str(-Z_critical) + " to " + str(Z_critical))

if Z_Stat > -Z_Critical:
    print("Z value falls in acceptance region")
    print("Accept Null Hypothesis at 5% significance level ")
else:
    print("Z value falls in rejection region")
    print("Reject Null Hypothesis at 5% significance level i.e., average dental expenses for the population is not accurate for the area")
    
x = np.arange(-4,4,0.001)
l =    str(-Z_critical) + " < Z < " +  str(Z_critical)
draw_z_score(x, (-Z_critical < x) & (x < Z_critical) , 0, 1, l,Z_Stat)
```

# Problem Statement 17:

A die is thrown 132 times with the following results:

Number turned up: 1, 2, 3, 4, 5, 6

Frequency: 16, 20, 25, 14, 29, 28

Is the die unbiased? Consider the degrees of freedom as n − 1.




# Solution:


H0: Die is unbaised.

H1: Die is baised.

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python

table = [[1,2,3,4,5,6],
         [16,20,25,14,29,28]]
print(table)
stat, p, dof, expected = chi2_contingency(table)
print('dof=%d' % dof)
print(expected)
# interpret test-statistic
prob = 0.95
critical = chi2.ppf(prob, dof)
print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
if abs(stat) >= critical:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
# interpret p-value
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Reject H0 - So the Die is biased')
else:
    print('Fail to reject H0 , so, Die is unbiased')
```

```python
table = [[18,22,20],
         [2,28,40],
        [20,10,40]]

print(table)
stat, p, dof, expected = chi2_contingency(table)
print('degrees of freedom=%d' % dof)
print(expected)
# interpret test-statistic
prob = 1 - 0.001
critical = chi2.ppf(prob, dof)
print('probability=%.3f, critical=%.3f, Chi_Square=%.3f' % (prob, critical, stat))
if abs(stat) >= critical:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
# interpret p-value
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Reject H0, hence, will accept that there is a significant relationship between age and photography. ')
else:
    print('Fail to reject H0, hence, will accept that there is no significant relationship between age and photography ')


```
